## Question 1

```python
import yfinance as yf
import pandas as pd

# here Creating a Ticker object
tesla_ticker = yf.Ticker("TSLA")

# now Extracting historical data
tesla_data = tesla_ticker.history(period="max")

# now Reseting the index
tesla_data.reset_index(inplace=True)

# lookiing for first 5 rows
tesla_data.head()
```

## Question 2

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Getting the HTML
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text
```

```python
# Parsing it using BeautifulSoup
soup = BeautifulSoup(html_data, "html.parser")
```

```python
# Using pandas to extract tables
tables = pd.read_html(html_data)
tesla_revenue = tables[1]  # The second table usually has the data we need
tesla_revenue.columns = ["Date", "Revenue"]
```

```python
# i will now clean the Revenue column
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(r"[\$,]", "", regex=True)
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors="coerce")
```

```python
# i also think dropping NA values will be a good idea to avoid discrepancy
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
```

## Question 5

```python
# for this better we convert into datetime
tesla_data["Date"] = pd.to_datetime(tesla_data["Date"])
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

# just to be safe we ensure Revenue is numeric
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors="coerce")
```

```python
import plotly.graph_objects as go

def make_graph(stock_data, revenue_data, stock):
    fig = go.Figure()

    # Stock Price Line Chart
    fig.add_trace(go.Scatter(x=stock_data["Date"], y=stock_data["Close"],
                             name="Stock Price"))

    # Revenue Bar Chart
    fig.add_trace(go.Scatter(x=revenue_data["Date"], y=revenue_data["Revenue"],
                             name="Revenue", yaxis="y2"))

    # Layout with two Y-axes
    fig.update_layout(
        title=stock + " Stock Price and Revenue",
        xaxis=dict(title="Date"),
        yaxis=dict(title="Stock Price"),
        yaxis2=dict(title="Revenue", overlaying="y", side="right"),
        legend=dict(x=0.1, y=1.1, orientation="h")
    )

    fig.show()
```

```python
make_graph(tesla_data, tesla_revenue, "Tesla")
```
